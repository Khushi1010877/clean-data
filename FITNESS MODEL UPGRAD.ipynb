```python
var['BookingEndDateTime (Month / Day / Year)'] = pd.to_datetime(var['BookingEndDateTime (Month / Day / Year)'], format='%d-%b-%y')
var['BookingStartTime'] = pd.to_datetime(var['BookingStartTime'], format='%H:%M:%S').dt.time

var['DayOfWeek'] = var['BookingEndDateTime (Month / Day / Year)'].dt.dayofweek
var['Month'] = var['BookingEndDateTime (Month / Day / Year)'].dt.month

var['Hour'] = pd.to_datetime(var['BookingStartTime'].astype(str)).dt.hour
```

```python
var['Price (INR)'] = pd.to_numeric(var['Price (INR)'], errors='coerce')
```

```python
var = var[var['Number Booked'] <= var['MaxBookees']]
```

```python
print(var.isnull().sum())
var = var.dropna()
```

```python
print(var.dtypes)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
min_max_scaler = MinMaxScaler()

# Columns to normalize
columns_to_normalize = ['MaxBookees', 'Number Booked']

# Apply normalization
var[columns_to_normalize] = min_max_scaler.fit_transform(var[columns_to_normalize])
```

```python
from sklearn.preprocessing import LabelEncoder

le_activity_site = LabelEncoder()
le_activity_desc = LabelEncoder()

var['ActivitySiteID_Num'] = le_activity_site.fit_transform(var['ActivitySiteID'])
#var['ActivityDescription_Num'] = le_activity_desc.fit_transform(var['ActivityDescription'])
var['Price (INR)'] = var['Price (INR)'].astype(int)
var
```

```python
var.drop('Month', axis=1, inplace=True)
```
